```python
mydf['away_fenwick_for'] = mydf['AwaySatf'] - mydf['HomeBlk']
mydf['away_fenwick_against'] = mydf['AwaySata'] - mydf['AwayBlk']

mydf['home_fenwick_for'] = mydf['HomeSatf'] - mydf['AwayBlk']
mydf['home_fenwick_against'] = mydf['HomeSata'] - mydf['HomeBlk']

mydf['away_fenwick_p'] = mydf['away_fenwick_for'] / (mydf['away_fenwick_against'] + mydf['away_fenwick_for'])
mydf['home_fenwick_p'] = mydf['home_fenwick_for'] / (mydf['home_fenwick_against'] + mydf['home_fenwick_for'])
```
