```python
from datetime import datetime, timedelta
#from datetime import timedelta

seasonStartBool = 1

teamStats = {}
offsetCSV = []
gameInsert = []
awayStats = None
homeStats = None
gameStats = None

gameInsertLive = []
offsetCSVLive = []

dateLimit = None
currentSeason = 0

for game in gameCSV:
    
    lastSeason = game[1] - 1
    currDate = datetime.strptime(game[0], '%Y-%m-%d')
    
#check to see if we have passed onto the next season
    if (seasonStartBool == 1) & (lastSeason == currentSeason):
        print("hit")
        currentSeason = 0
        
#set the datelimit for the currentSeason
    if (seasonStartBool == 1) & (currentSeason == 0):
        date = datetime.strptime(game[0], '%Y-%m-%d')
        dateForward = date + timedelta(days = 17)
        dateLimit = dateForward
        currentSeason = game[1]
        print(game)
    
    
    if currDate < dateLimit:
        if seasonStartBool == 1:
            teamStats = {}
            seasonStartBool = 0
        
        awayStats = game[4:39]
        homeStats = game[39:74]
        
        awaySRS = game[38]
        homeSRS = game[73]
        
        gameDate = game[0:2]

        if game[2] not in teamStats.keys():
            teamStats[game[2]] = awayStats
            teamStats[game[2]].append(1)
            
#if game already exists in the dicitionary add 1 to the number of games played(last item) then pull the list of stats from the dictionary
#loop through the all the team stats except the last one since we handle that outside the for loop first
#and identify the index of each item to pull the corresponding position of the stat from the parent if then statemetn above where we initially define the stats of the current row
#this logic will be used later as well when we update the dictionary stats after populating offset csv


#need to remember that the calculation of per game stats is incredibly difficult because the dictionary will contain
#per game stats but we only have single game stats in a given week. thus we need to mulitply the value in the dictionary by the games played
#thus far to get to a total for the season, add the current week value, then divide by the new total games played

        elif game[2] in teamStats.keys():
            gamesPlayedOriginal = teamStats[game[2]][-1]
            gamesPlayedUpdate = gamesPlayedOriginal + 1
            teamStats[game[2]][-1] = gamesPlayedUpdate

#operation stats is the most up to date copy of the stats from the dictionary which we then update before reassignig the dictionary key
            operationStats = teamStats[game[2]]
            updateList = []
            indexCounter = 0
#-2 because SRS is the second to last stat and is not an average
            for item in operationStats[:-2]:
                newStat = round(((float(item)*float(gamesPlayedOriginal)) + float(awayStats[indexCounter])) / float(gamesPlayedUpdate),4)
                updateList.append(newStat)
                indexCounter += 1
            
            teamStats[game[2]] = updateList
            teamStats[game[2]].append(awaySRS)
            teamStats[game[2]].append(gamesPlayedUpdate)
        
        if game[3] not in teamStats.keys():
            teamStats[game[3]] = homeStats
            teamStats[game[3]].append(1)
            
        elif game[3] in teamStats.keys():
            gamesPlayedOriginal = teamStats[game[3]][-1]
            gamesPlayedUpdate = gamesPlayedOriginal + 1
            teamStats[game[3]][-1] = gamesPlayedUpdate

#operation stats is the most up to date copy of the stats from the dictionary which we then update before reassignig the dictionary key
            operationStats = teamStats[game[3]]
            updateList = []
            indexCounter = 0
            for item in operationStats[:-2]:
                newStat = round(((float(item*gamesPlayedOriginal)) + float(homeStats[indexCounter])) / float(gamesPlayedUpdate),4)
                updateList.append(newStat)
                indexCounter += 1
                
            teamStats[game[3]] = updateList
            teamStats[game[3]].append(homeSRS)
            teamStats[game[3]].append(gamesPlayedUpdate)

    else:
        if currDate >= dateLimit:
            seasonStartBool = 1
# #make the offsetCSV with the most up to date stats first
# #making the decision here to exclude week, date, day, everything up to team1 and team2 above in the final offset dataset
        awayStats = game[4:39]
        homeStats = game[39:74]
    
        matchup = game[2:4]
        gameStats = game[74:]
        
        awayKey = game[2]
        homeKey = game[3]
        
        awaySRS = game[38]
        homeSRS = game[73]
        
        gameDate = game[0:2]
#currently, if a team hasnt played by week 3, or an unencountered team is appearing, we are just setting stats to nan to later be dropped
#what would be more interesting is if we find the average team stats at all times and apply instead
        try:
            awayStatsLookup = teamStats[awayKey]
        except:
            awayStatsLookup = [numpy.nan] * 29
        try:
            homeStatsLookup = teamStats[homeKey]
        except:
            homeStatsLookup = [numpy.nan] * 29
#here we are adding the matchup, the stats from the dictionary from team 1 and 2, and the game stats
        gameInsert.extend(gameDate)
        gameInsert.extend(matchup)
        gameInsert.extend(awayStatsLookup)
        gameInsert.extend(homeStatsLookup)
        gameInsert.extend(gameStats)
#update the dictionary
#copy the logic from above, if a team for some reason doesnt play for the first two weeks add them in, else update the stats
        if game[2] not in teamStats.keys():
            teamStats[game[2]] = awayStats
            teamStats[game[2]].append(1)
            
        elif game[2] in teamStats.keys():
            gamesPlayedOriginal = teamStats[game[2]][-1]
            gamesPlayedUpdate = gamesPlayedOriginal + 1
            teamStats[game[2]][-1] = gamesPlayedUpdate

            operationStats = teamStats[game[2]]
            updateList = []
            indexCounter = 0
            for item in operationStats[:-2]:
#think very carefully here about which stats should be per game average and which should not, at first glance I think they all can be
                newStat = round(((float(item)*float(gamesPlayedOriginal)) + float(awayStats[indexCounter]))/ float(gamesPlayedUpdate), 4)
                updateList.append(newStat)
                indexCounter += 1
            
            teamStats[game[2]] = updateList
            teamStats[game[2]].append(awaySRS)
            teamStats[game[2]].append(gamesPlayedUpdate)
        
        if game[3] not in teamStats.keys():

            teamStats[game[3]] = homeStats
            teamStats[game[3]].append(1)
            
        elif game[3] in teamStats.keys():
            gamesPlayedOriginal = teamStats[game[3]][-1]
            gamesPlayedUpdate = gamesPlayedOriginal + 1
            teamStats[game[3]][-1] = gamesPlayedUpdate

#operation stats is the most up to date copy of the stats from the dictionary which we then update before reassignig the dictionary key
            operationStats = teamStats[game[3]]
            updateList = []
            indexCounter = 0
            for item in operationStats[:-2]:
#think very carefully here about which stats should be per game average and which should not, at first glance I think they all can be
                newStat = round(((float(item)*float(gamesPlayedOriginal)) + float(homeStats[indexCounter]))/ float(gamesPlayedUpdate), 4)
                updateList.append(newStat)
                indexCounter += 1
            
            teamStats[game[3]] = updateList
            teamStats[game[3]].append(homeSRS)
            teamStats[game[3]].append(gamesPlayedUpdate)

        offsetCSV.append(gameInsert)
        gameInsert = []
        
        #here we are adding the matchup, the stats from the dictionary from team 1 and 2, and the game stats
        gameInsertLive.extend(gameDate)
        gameInsertLive.extend(matchup)
        gameInsertLive.extend(awayStatsLookup)
        gameInsertLive.extend(homeStatsLookup)
        gameInsertLive.extend(gameStats)
        
        offsetCSVLive.append(gameInsertLive)
        gameInsertLive = []
```
