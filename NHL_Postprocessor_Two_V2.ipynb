```python
oddsdf = oddsdf[[
    'DateClean',
    'Season',
#    'VH_Away',
    'Away',
#    'V_1st',
#    'V_2nd',
#    'V_3rd',
    'V_Final',
    'V_Open',
    'V_Close',
#    'V_Spread',
#    'V_PuckLine',
#    'V_OpenOU',
#    'V_OpenOUSpread',
#    'V_CloseOU',
#    'V_CloseOUSpread',
#    'VH_Home',
    'Home',
#    'H_1st',
#    'H_2nd',
#    'H_3rd', 
    'H_Final',
    'H_Open',
    'H_Close',
#    'H_Spread',
#    'H_PuckLine',
#    'H_OpenOU',
#    'H_OpenOUSpread',
#    'H_CloseOU',
#    'H_CloseOUSpread',
    'JoinID'
]]
```

```python
mydf2 = pandas.merge(left = mydf, right = oddsdf, left_on = 'JoinID',right_on = 'JoinID', how = 'inner')
```

```python
cols = mydf2.columns.tolist()
```

```python
import numpy as np
mydf2['OpenHomeIPW'] = np.where(mydf2['H_Open'] < 0, (1/((100 / abs(mydf2['H_Open'])) + 1)), (1/((mydf2['H_Open'] / 100) + 1)))
mydf2['OpenAwayIPW'] = np.where(mydf2['V_Open'] < 0, (1/((100 / abs(mydf2['V_Open'])) + 1)), (1/((mydf2['V_Open'] / 100) + 1)))

mydf2['CloseHomeIPW'] = np.where(mydf2['H_Close'] < 0, (1/((100 / abs(mydf2['H_Close'])) + 1)), (1/((mydf2['H_Close'] / 100) + 1)))
mydf2['CloseAwayIPW'] = np.where(mydf2['V_Close'] < 0, (1/((100 / abs(mydf2['V_Close'])) + 1)), (1/((mydf2['V_Close'] / 100) + 1)))

mydf2['HomeIPVMove'] = mydf2['CloseHomeIPW'] - mydf2['OpenHomeIPW']
mydf2['AwayIPVMove'] = mydf2['CloseAwayIPW'] - mydf2['OpenAwayIPW']
```

```python
mydf2
```

```python
mydf2['vegasPredIPW'] = numpy.where(mydf2['OpenHomeIPW'] > mydf2['OpenAwayIPW'], mydf2['OpenHomeIPW'], mydf2['OpenAwayIPW'])
filterdf = mydf2.loc[(mydf2['vegasPredIPW'] >= 0.5) & (mydf2['vegasPredIPW'] < 1)].copy()

filterdf['vegasPred'] = numpy.where(filterdf['OpenHomeIPW'] > filterdf['OpenAwayIPW'], 'Home', 'Away')
filterdf['winner'] = numpy.where(filterdf['HomeScore'] > filterdf['AwayScore'], 'Home', 'Away')
filterdf['vegasOpenCorrect'] = numpy.where(filterdf['vegasPred'] ==  filterdf['winner'], 1, 0)
vegasOpenCorrect = filterdf['vegasOpenCorrect'].sum()
totalGames = len(filterdf)


print(filterdf['vegasOpenCorrect'].sum())
print(len(filterdf))
print(vegasOpenCorrect/totalGames)
```

```python
mydf2['vegasPredIPW'] = numpy.where(mydf2['CloseHomeIPW'] > mydf2['CloseAwayIPW'], mydf2['CloseHomeIPW'], mydf2['CloseAwayIPW'])
filterdf = mydf2.loc[(mydf2['vegasPredIPW'] >= 0.6) & (mydf2['vegasPredIPW'] < 0.65)].copy()

filterdf['vegasPred'] = numpy.where(filterdf['CloseHomeIPW'] > filterdf['CloseAwayIPW'], 'Home', 'Away')
filterdf['winner'] = numpy.where(filterdf['HomeScore'] > filterdf['AwayScore'], 'Home', 'Away')
filterdf['vegasCloseCorrect'] = numpy.where(filterdf['vegasPred'] ==  filterdf['winner'], 1, 0)
vegasCloseCorrect = filterdf['vegasCloseCorrect'].sum()
totalGames = len(filterdf)


print(filterdf['vegasCloseCorrect'].sum())
print(len(filterdf))
print(vegasCloseCorrect/totalGames)
```
