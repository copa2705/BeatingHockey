```python
import datetime

elodf['date'].astype('datetime64[ns]')
elodf['date'] =  pandas.to_datetime(elodf['date'], format='%m/%d/%Y')
elodf['date'] = elodf['date'] + datetime.timedelta(days = 1)
elodf['date'] = elodf['date'].dt.strftime('%Y-%m-%d')
elodf['date'].astype('str')
elodf.insert(0, 'joinID', value = elodf['date'] + elodf['home_team'])

elodf['season'].astype('int')
elodf2 = elodf.loc[elodf['season'] >= 2009]
elodf3 = elodf2.loc[elodf2['season'] <= 2020]
elodf4 = elodf3.loc[elodf3['playoff'] != 1]
```
