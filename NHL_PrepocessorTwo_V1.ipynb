## initialize the Margin of Victory which is the rating used in SRS systems

In SRS, a teams rating is their average margin of victory + the weighted average of the opponents they have played
Thus, we have a system of equations which once solved yields the rating of a team adjusted for the strength of their opponents

```python

homeAdjustment = 0.25

df2['AwayMov'] = df2['AwayScore'] - df2['HomeScore'] + homeAdjustment
df2['HomeMov'] = df2['HomeScore'] - df2['AwayScore'] - homeAdjustment


df2.head(10)
    
```

currently have a weekly generation of SRS by team including that weeks results. now I need to reconstruct the original dataframe with joins off the SRS dataframe and match the SRS by team name. Note that weeks 1-3 will not have SRS. set these missing to 0. this means when offsetting, starting in week 4 we will have the week 3 results which will need to be included, then the team dictionary will need to be updated with the new SRS values after the week 4 action.

need to find an alternative to the week system...should be solved every time there is a change in date so unique date in a given season

```python
df4 = pandas.DataFrame()
seasonList = df2['Season'].unique()
seasons = seasonList.tolist()
print(seasons)
```

```python
for year in seasons:
    print(year)
#loop through list of seasons and cut out the yearly dataframes
    seasonDf = df2.loc[df2.Season == year]

#identify unique number of Dates
    dayRaw= seasonDf['DateClean'].unique()
    days = dayRaw.tolist()
                                                                                  
# #loop through the list of weeks and cut out the weekly dataframes
    for day in days:
#        print(day)
        
#so for week through, we take all games from week 1 through 4 inclusive
        dayDf = seasonDf.loc[seasonDf.DateClean <= day]

#split the data into two rows for each game, team 1 renamed to team, team1 Margin of Victory (MOV) renamed to spread, team 2 renamed to opponent
#vice versa for team 2
        dayTeams = pandas.concat([
        dayDf[['Away', 'AwayMov', 'Home']].rename(columns={'Away': 'Team', 'AwayMov': 'spread', 'Home': 'Opponent'}),
        dayDf[['Home', 'HomeMov', 'Away']].rename(columns={'Home': 'Team', 'HomeMov': 'spread', 'Away': 'Opponent'})
        ])
#        print(dayTeams)
#cap the wins to 4 to eliminate noise from blowouts
        dayTeams['spread'] = np.where(dayTeams['spread'] > 4, 4, dayTeams['spread']) # cap the upper bound scoring margin at +28 points
        dayTeams['spread'] = np.where(dayTeams['spread'] < -4, -4, dayTeams['spread']) # cap the lower bound scoring margin at -28 points

#create a dataframe grouped by team with the average margin of victory
        daySpreads = dayTeams.groupby('Team').spread.mean()

# create empty arrays for linear algebra, the equations and the solution matrices
        terms = []
        solutions = []

        for team in daySpreads.keys():
#            print(team)
            row = []
# get a list of team opponents
            opps = list(dayTeams[dayTeams['Team'] == team]['Opponent'])
#            print(opps)

            for opp in daySpreads.keys():
                if opp == team:
# coefficient for the team should be 1
                    row.append(1)
                elif opp in opps:
# coefficient for opponents should be 1 over the number of opponents
                    row.append(-1.0/len(opps))
                else:
                    # teams not faced get a coefficient of 0
                    row.append(0)

            terms.append(row)

# average game spread on the other side of the equation
            solutions.append(daySpreads[team])
#        print(terms)

        try:
#try to solve the system of equation
            daySolutions = np.linalg.solve(np.array(terms), np.array(solutions))
#create a list of lists containing the team names and the solutions created above
            dayRatings = list(zip( daySpreads.keys(), solutions ))
#create dataframe of list
            srs = pandas.DataFrame(dayRatings, columns=['team', 'rating'])
            srsValues = srs.set_index('team')['rating'].to_dict()
#map the weekly SRS values as of that week to the srs fields using team names as keys
            dayDf2 = dayDf.copy(deep = True)
            dayDf2['AwaySRS'] = dayDf['Away'].map(srsValues)
            dayDf2['HomeSRS'] = dayDf['Home'].map(srsValues)
#cutoff the old weeks data, leaving the most current week
            dayDf3 = dayDf2.loc[dayDf2.DateClean == day]
        except:
#if the system of equations is unsolvable (when a team hasnt played yet or some nonsense) set SRS to na (later 0) and move on
            print("error" + str(year) + ", " + str(day))
            dayDf2 = dayDf.copy(deep = True)
            dayDf2['AwaySRS'] = numpy.nan
            dayDf2['HomeSRS'] = numpy.nan
            dayDf3 = dayDf2.loc[dayDf2.DateClean == day]
#finally add the JUST the new week of data to the dataframe 
        df4 = pandas.concat([df4, dayDf3])
#print(df4)




    
    
```
